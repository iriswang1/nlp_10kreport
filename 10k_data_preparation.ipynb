### Add industry variable into the dataset 
 - To improve the model performance

```python
def readFile(file_path, main_path):
    
    full_path = main_path + file_path
    
    with open(full_path, 'r') as file:
        
        file_content = file.read()
        
    return file_content

def getIndustry(file_path, main_path):
    
    file_content = readFile(file_path, main_path)

    rf_1 = re.search(r'STANDARD[\s]*INDUSTRIAL[\s]*CLASSIFICATION:(.{30,}?)IRS[\s]*NUMBER[\s]*:', file_content, re.S)

    if rf_1:
        text_industry = re.sub(r'\s{2,}',' ',rf_1.group())
        return text_industry

    
```

```python
#get all risk facts for every file

main_path = '/Users/iriswang/Google Drive/Stevens/Fall 2018/BIA660/Project/Data/10-K Plain Text/edgar_text/'

index_kmv_data['industry'] = index_kmv_data["actual_10k_path_raw"].apply(
    lambda file_path: getIndustry(file_path, main_path))

index_kmv_data.head()
```
